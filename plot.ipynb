```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

# データ
P = [8, 16, 32, 64, 128, 256]
psnr = [26.44, 26.43, 26.44, 26.47, 26.47, 26.44]
ssim = [0.7373, 0.7365, 0.7368, 0.7415, 0.7375, 0.7385]

# プロット作成
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# PSNRプロット（柔らかいけど濃い赤）
ax1.plot(range(len(P)), psnr, marker='o', color="#FF6666", label="PSNR", linewidth=2.5)
# SSIMプロット（柔らかいけど濃い青）
ax2.plot(range(len(P)), ssim, marker='s', color="#6699CC", label="SSIM", linewidth=2.5)

# 横軸目盛りをPに従う
ax1.set_xticks(range(len(P)))
ax1.set_xticklabels(P)

# PSNRのy軸設定
ax1.set_ylabel("PSNR", fontsize=12)
ax1.yaxis.set_major_locator(MultipleLocator(0.01))  # 主目盛りを0.01刻み
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))  # 小数点以下2桁にフォーマット

# SSIMのy軸設定
ax2.set_ylabel("SSIM", fontsize=12)
ax2.yaxis.set_major_locator(MultipleLocator(0.001))  # 主目盛りを0.001刻み
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.4f'))  # 小数点以下4桁にフォーマット

# 凡例の設定（右下に配置）
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1 + handler2, label1 + label2, loc='lower right', fontsize=12)

# グリッドを追加
ax1.grid(which='both', linestyle='--', linewidth=0.7, alpha=0.7)

# 軸ラベル
ax1.set_xlabel("Number of filters (P)", fontsize=12)

# レイアウト調整
fig.tight_layout()

# プロット表示
plt.show()
```
