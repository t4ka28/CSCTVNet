```python
import matplotlib.pyplot as plt

# データ
k = [3, 5, 7, 9, 11]
psnr = [26.1, 26.29, 26.39, 26.36, 26.3]
ssim = [0.7211, 0.7286, 0.7356, 0.7365, 0.7324]

# プロット作成
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# PSNRプロット（柔らかいけど濃い赤）
ax1.plot(k, psnr, marker='o', color="#FF6666", label="PSNR", linewidth=2.5)
# SSIMプロット（柔らかいけど濃い青）
ax2.plot(k, ssim, marker='s', color="#6699CC", label="SSIM", linewidth=2.5)

# 凡例の設定（右下に配置）
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1 + handler2, label1 + label2, loc='lower right', fontsize=12)

# 軸ラベルの設定
ax1.set_xlabel("Kernel Size")
ax1.set_ylabel("PSNR")  # PSNR軸ラベルを濃い赤
ax2.set_ylabel("SSIM")  # SSIM軸ラベルを濃い青

# 横軸目盛りをkに従う
ax1.set_xticks(k)

# グリッドを追加
ax1.grid(which='both', linestyle='--', linewidth=0.7)

# レイアウト調整
fig.tight_layout()

# プロット表示
plt.show()
```

```python
import matplotlib.pyplot as plt

# データ
k = [8, 16, 32, 64, 128]
psnr = [26.1, 26.29, 26.39, 26.36, 26.3]
ssim = [0.7211, 0.7286, 0.7356, 0.7365, 0.7324]

# プロット作成
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# PSNRプロット（柔らかい赤）
ax1.plot(range(len(k)), psnr, marker='o', color="#FF6666", label="PSNR", linewidth=2.5)
# SSIMプロット（柔らかい青）
ax2.plot(range(len(k)), ssim, marker='s', color="#6699CC", label="SSIM", linewidth=2.5)

# x軸のラベルを等間隔で配置
ax1.set_xticks(range(len(k)))
ax1.set_xticklabels(k)

# 凡例の設定（右下に配置、フォントサイズ大きめ）
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1 + handler2, label1 + label2, loc='lower right', fontsize=12)

# 軸ラベルの設定
ax1.set_xlabel("k (Kernel Size)")
ax1.set_ylabel("PSNR", color="#CC3333")  # PSNR軸ラベルを濃い赤
ax2.set_ylabel("SSIM", color="#336699")  # SSIM軸ラベルを濃い青

# グリッドを追加
ax1.grid(which='both', linestyle='--', linewidth=0.7)

# レイアウト調整
fig.tight_layout()

# プロット表示
plt.show()
```
